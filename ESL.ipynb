```python
class Base:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    
    def info(self):
        return ('Name:{} Age:{}'.format(self.name,self.age))
```

```python
class Der(Base):
    def __init__(self,name,age,score):
        super().__init__(name,age)
        self.score = score
    
    def info(self):
        return ('Name:{} Age:{} Score:{}'.format(self.name,self.age,self.score))
```

```python
der = Der('Bob','12','50')
der.info()
```
